```python
def primesUpTo(number):
    prime_list = []
    for number in range(2, number + 1):
        primality = True
        if (number % 2 != 0) or (number == 2):
            for divisor in range(3, int(number**(1/2)) + 1, 2):
                if (number % divisor) == 0:
                    primality = False
        else:
            primality = False
        if primality == True:
            prime_list.append(number)
    return sum(prime_list)

primesUpTo(2000000)
```
